A. Función de Resumen de Calidad de Datos
Esta función ayuda a identificar rápidamente tipos de datos, nulos y el porcentaje de completitud.

```python
import pandas as pd

url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df = pd.read_csv(url)

# Para comprobar que cargó bien
print(df.head())
def describe_df(df):
    """
    Genera un resumen del dataframe: tipo, nulos y % de nulos.
    """
    summary = pd.DataFrame({
        'Tipo': df.dtypes,
        'Nulos': df.isnull().sum(),
        '% Nulos': (df.isnull().sum() / len(df)) * 100
    })
    return summary.sort_values(by='% Nulos', ascending=False)
    
```

```python
# 1. Ver información básica
print(df.info())

# 2. Aplicar la función de la ToolBox que describe nulos (si ya está creada)
resumen = describe_df(df)
print(resumen)

# 3. Probar la función de visualización
plot_combined_grid(df, 'Age')
```

B. Función de Visualización Automática (Matplotlib)
Siguiendo la recomendación de usar Matplotlib, puedes crear una función que grafique la distribución de una variable numérica:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df = pd.read_csv(url)

# 2. DEFINIR LA FUNCIÓN (La "receta")
def plot_combined_grid(df, column):
    # Creamos la estructura del gráfico
    fig, (ax_box, ax_hist) = plt.subplots(2, sharex=True, 
                                          gridspec_kw={"height_ratios": (.15, .85)},
                                          figsize=(10, 6))
    
    # Añadimos el Boxplot arriba
    sns.boxplot(x=df[column], ax=ax_box, color="skyblue")
    ax_box.set(title=f'Distribución de: {column}', xlabel='')

    # Añadimos el Histograma abajo
    sns.histplot(df[column], kde=True, ax=ax_hist, color="skyblue")
    
    plt.tight_layout()
    plt.show() # <--- ¡IMPORTANTE! Esto le dice a Python: "Muéstrame el dibujo"

# 3. EJECUTAR LA FUNCIÓN 
plot_combined_grid(df, 'Age')
```
